# What months or seasons have higher delays?

## Delays by months

```python
df['FL_MONTH_NAME']
```

```python
delays_per_month.plot(kind='barh')
```

## Delays by seasons

```python
def get_season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    else:
        return 'Fall'

df['FL_SEASON'] = df['FL_MONTH_NAME'].apply(get_season)
df['FL_SEASON']
```

```python
Delays_by_seasons = df[df['ARR_DELAY'] > 0].groupby('FL_SEASON')['ARR_DELAY'].count().sort_values(ascending=True)
Delays_by_seasons
```

```python
Delays_by_seasons.plot(kind='barh')
```

```python
seasonal_delay_by_cause = df.groupby('FL_SEASON').agg({
    'DELAY_DUE_CARRIER': lambda x: (x > 0).sum(),
    'DELAY_DUE_NAS': lambda x: (x > 0).sum(),
    'DELAY_DUE_WEATHER': lambda x: (x > 0).sum(),
    'DELAY_DUE_SECURITY': lambda x: (x > 0).sum(),
})
seasonal_delay_by_cause.plot(kind='line')
plt.xlabel('')
plt.legend(loc= 'upper left')
plt.show()
```

# Which delay cause (carrier, weather, NAS, Security) dominates?

# Count of delay by each Airline per cause

```python
count_of_delay_cause = df.groupby('AIRLINE').agg({
    'DELAY_DUE_CARRIER': lambda x: (x > 0).sum(),
    'DELAY_DUE_WEATHER': lambda x: (x > 0).sum(),
    'DELAY_DUE_NAS': lambda x: (x > 0).sum(),
    'DELAY_DUE_SECURITY': lambda x: (x > 0).sum()
}).sort_values(by='DELAY_DUE_CARRIER',ascending=True)
count_of_delay_cause = count_of_delay_cause.reset_index()
count_of_delay_cause
```

# All in one

```python
fig, ax = plt.subplots(2,2, figsize = (16,9))
count_of_delay_cause.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_CARRIER', ax=ax[0,0])
count_of_delay_cause.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_NAS', ax=ax[0,1])
count_of_delay_cause.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_WEATHER', ax=ax[1,0])
count_of_delay_cause.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_SECURITY', ax=ax[1,1])
ax[0,0].set_title('Flights delayed due to carrier')
ax[0,1].set_title('Flights delayed due to Air traffic')
ax[1,0].set_title('Flights delayed due to Weather')
ax[1,1].set_title('Flights delayed due to Security')

fig.tight_layout()
```

# Average delay time of each Airline per cause

```python
avg_of_delays = df.groupby('AIRLINE').agg({
    'DELAY_DUE_CARRIER': lambda x: x[x > 0].mean(),
    'DELAY_DUE_WEATHER': lambda x: x[x > 0].mean(),
    'DELAY_DUE_NAS': lambda x: x[x > 0].mean(),
    'DELAY_DUE_SECURITY' : lambda x: x[x > 0].mean()
}).sort_values(by='DELAY_DUE_CARRIER',ascending=True).reset_index()
avg_of_delays
```

# All in one

```python
fig, ax = plt.subplots(2,2, figsize = (16,9))
avg_of_delays.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_CARRIER', ax=ax[0,0])
avg_of_delays.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_NAS', ax=ax[0,1])
avg_of_delays.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_WEATHER', ax=ax[1,0])
avg_of_delays.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_SECURITY', ax=ax[1,1])
ax[0,0].set_title('Average Flights delayed time due to carrier')
ax[0,1].set_title('Average Flights delayed time due to Air traffic')
ax[1,0].set_title('Average Flights delayed time due to Weather')
ax[1,1].set_title('Average Flights delayed time due to Security')
fig.tight_layout()
```

# Sum of delay time by each Airline per cause

```python
sum_of_delay_cause = df.groupby('AIRLINE').agg({
    'DELAY_DUE_CARRIER': 'sum',
    'DELAY_DUE_WEATHER': 'sum',
    'DELAY_DUE_NAS': 'sum',
    'DELAY_DUE_SECURITY': 'sum'
}).sort_values(by='DELAY_DUE_CARRIER',ascending=True).reset_index()
sum_of_delay_cause
```

# All in one

```python
fig, ax = plt.subplots(2,2, figsize = (16,9))
sum_of_delay_cause.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_CARRIER', ax=ax[0,0])
sum_of_delay_cause.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_NAS', ax=ax[0,1])
sum_of_delay_cause.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_WEATHER', ax=ax[1,0])
sum_of_delay_cause.plot(kind='barh', x = 'AIRLINE', y='DELAY_DUE_SECURITY', ax=ax[1,1])
ax[0,0].set_title('Total Flights delayed time due to carrier in minutes')
ax[0,1].set_title('Total Flights delayed time due to Air traffic in minutes')
ax[1,0].set_title('Total Flights delayed time due to Weather in minutes')
ax[1,1].set_title('Total Flights delayed time due to Security in minutes')
fig.tight_layout()
```
